```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import xgboost as xgb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
import seaborn as sns
import logging
from typing import Dict, List, Any, Union, Optional
from pathlib import Path
```

```python
# 設定數據路徑
DATA_PATH = Path("D:/Min/Python/Project/FA_Data")
META_DATA_PATH = DATA_PATH / "meta_data"

def load_stock_data(stock_id: str, start_date: str, end_date: str) -> pd.DataFrame:
    """
    載入股票數據和特徵
    
    Args:
        stock_id: 股票代碼
        start_date: 起始日期 (YYYY-MM-DD)
        end_date: 結束日期 (YYYY-MM-DD)
        
    Returns:
        包含原始數據和特徵的DataFrame
    """
    try:
        # 1. 讀取原始價格數據
        base_dir = Path("D:/Min/Python/Project/FA_Data")
        meta_data_path = base_dir / "meta_data" / "all_stocks_data.csv"
        
        logging.info(f"正在讀取原始數據: {meta_data_path}")
        price_df = pd.read_csv(meta_data_path, low_memory=False)
        
        # 過濾股票和日期
        price_df = price_df[price_df['證券代號'] == stock_id]
        price_df['日期'] = pd.to_datetime(price_df['日期'])
        price_df = price_df[(price_df['日期'] >= start_date) & (price_df['日期'] <= end_date)]
        
        # 2. 讀取特徵數據
        feature_file = base_dir / "features" / f"{stock_id}_features_{start_date}_{end_date}.csv"
        
        if not feature_file.exists():
            logging.error(f"特徵文件不存在: {feature_file}")
            logging.info("請先運行 03_feature_generator.ipynb 生成特徵")
            raise FileNotFoundError(f"找不到特徵文件: {feature_file}")
        
        logging.info(f"正在讀取特徵數據: {feature_file}")
        feature_df = pd.read_csv(feature_file)
        
        # 確保日期格式一致
        feature_df['日期'] = pd.to_datetime(feature_df['日期'])
        
        # 3. 合併數據
        # 保留需要的原始列
        price_cols = ['證券代號', '證券名稱', '日期', '收盤價', '成交金額', '成交股數']
        df = price_df[price_cols].merge(feature_df, on='日期', how='inner')
        
        logging.info(f"數據載入完成，形狀: {df.shape}")
        
        return df
        
    except Exception as e:
        logging.error(f"載入數據時出錯: {str(e)}")
        raise
```

```python
class BaseModel:
    """模型基礎類"""
    
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        """訓練模型"""
        raise NotImplementedError
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        """預測"""
        raise NotImplementedError
    
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """預測概率"""
        raise NotImplementedError

class RandomForestModel(BaseModel):
    """隨機森林模型"""
    
    def __init__(self, params: Optional[Dict] = None):
        default_params = {
            'n_estimators': 100,
            'max_depth': 5,
            'random_state': 42
        }
        self.params = default_params if params is None else {**default_params, **params}
        self.model = RandomForestClassifier(**self.params)
    
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        self.model.fit(X, y)
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict(X)
    
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict_proba(X)

class XGBoostModel(BaseModel):
    """XGBoost模型"""
    
    def __init__(self, params: Optional[Dict] = None):
        default_params = {
            'max_depth': 6,
            'learning_rate': 0.1,
            'objective': 'binary:logistic',
            'random_state': 42
        }
        self.params = default_params if params is None else {**default_params, **params}
        self.model = xgb.XGBClassifier(**self.params)
    
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        self.model.fit(X, y)
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict(X)
    
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict_proba(X)

class LSTMModel(BaseModel):
    """LSTM模型"""
    
    def __init__(self, params: Optional[Dict] = None):
        default_params = {
            'units': 50,
            'dropout': 0.2,
            'epochs': 100,
            'batch_size': 32
        }
        self.params = default_params if params is None else {**default_params, **params}
        self.model = self._build_model()
    
    def _build_model(self) -> Sequential:
        model = Sequential([
            LSTM(self.params['units'], return_sequences=True),
            Dropout(self.params['dropout']),
            LSTM(self.params['units']//2),
            Dropout(self.params['dropout']),
            Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model
    
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        if len(X.shape) == 2:
            X = X.reshape((X.shape[0], 1, X.shape[1]))
        self.model.fit(
            X, y,
            epochs=self.params['epochs'],
            batch_size=self.params['batch_size'],
            verbose=0
        )
    
    def predict(self, X: np.ndarray) -> np.ndarray:
        if len(X.shape) == 2:
            X = X.reshape((X.shape[0], 1, X.shape[1]))
        return (self.model.predict(X) > 0.5).astype(int)
    
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        if len(X.shape) == 2:
            X = X.reshape((X.shape[0], 1, X.shape[1]))
        probs = self.model.predict(X)
        return np.hstack((1-probs, probs))
```

```python
class ModelFactory:
    """模型工廠類"""
    
    @staticmethod
    def create_model(model_type: str, params: dict = None) -> BaseModel:
        """
        創建模型實例
        
        Args:
            model_type: 模型類型 ("random_forest", "xgboost", "lstm")
            params: 模型參數字典
            
        Returns:
            模型實例
        """
        if model_type == "random_forest":
            return RandomForestModel(params)
        elif model_type == "xgboost":
            return XGBoostModel(params)
        elif model_type == "lstm":
            return LSTMModel(params)
        else:
            raise ValueError(f"不支援的模型類型: {model_type}")

class ModelConfig:
    """模型配置類"""
    
    # 隨機森林默認參數
    RF_PARAMS = {
        'n_estimators': 100,
        'max_depth': 5,
        'min_samples_split': 2,
        'min_samples_leaf': 1,
        'random_state': 42
    }
    
    # XGBoost默認參數
    XGB_PARAMS = {
        'max_depth': 6,
        'learning_rate': 0.1,
        'n_estimators': 100,
        'objective': 'binary:logistic',
        'random_state': 42
    }
    
    # LSTM默認參數
    LSTM_PARAMS = {
        'units': 50,
        'dropout': 0.2,
        'epochs': 100,
        'batch_size': 32
    }
    
    # 參數優化搜索範圍
    RF_PARAM_GRID = {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 5, 7],
        'min_samples_split': [2, 5, 10]
    }
    
    XGB_PARAM_GRID = {
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.1, 0.3],
        'n_estimators': [100, 200, 300]
    }
```

```python
class ModelEvaluator:
    """模型評估器"""
    
    @staticmethod
    def evaluate_classification(y_true: np.ndarray, 
                              y_pred: np.ndarray, 
                              y_prob: np.ndarray = None) -> Dict[str, float]:
        """
        評估分類模型效果
        
        Args:
            y_true: 真實標籤
            y_pred: 預測標籤
            y_prob: 預測概率
            
        Returns:
            包含各項評估指標的字典
        """
        metrics = {
            'accuracy': accuracy_score(y_true, y_pred),
            'precision': precision_score(y_true, y_pred),
            'recall': recall_score(y_true, y_pred),
            'f1': f1_score(y_true, y_pred)
        }
        
        # 如果有概率預測，計算AUC
        if y_prob is not None:
            metrics['auc'] = roc_auc_score(y_true, y_prob[:, 1])
            
        return metrics
    
    @staticmethod
    def print_evaluation_report(metrics: Dict[str, float], model_name: str = "Model") -> None:
        """打印評估報告"""
        print(f"\n{model_name} Evaluation Report:")
        print("-" * 40)
        for metric, value in metrics.items():
            print(f"{metric.capitalize():15}: {value:.4f}")
    
    @staticmethod
    def plot_confusion_matrix(y_true: np.ndarray, 
                            y_pred: np.ndarray, 
                            title: str = "Confusion Matrix") -> None:
        """繪製混淆矩陣"""
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(title)
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()
```

```python
class ModelOptimizer:
    """模型優化器"""
    
    def __init__(self, model_type: str, param_grid: Dict[str, List[Any]]):
        self.model_type = model_type
        self.param_grid = param_grid
        self.best_params = None
        self.best_score = None
        
    def optimize(self, X: np.ndarray, y: np.ndarray, 
                cv: int = 5, scoring: str = 'accuracy') -> Dict[str, Any]:
        """
        執行參數優化
        
        Args:
            X: 特徵矩陣
            y: 標籤向量
            cv: 交叉驗證折數
            scoring: 評分標準
            
        Returns:
            優化結果字典
        """
        logging.info(f"開始對 {self.model_type} 進行參數優化...")
        
        grid_search = GridSearchCV(
            ModelFactory.create_model(self.model_type).model,
            self.param_grid,
            cv=cv,
            scoring=scoring,
            n_jobs=-1,
            verbose=1
        )
        
        grid_search.fit(X, y)
        
        self.best_params = grid_search.best_params_
        self.best_score = grid_search.best_score_
        
        logging.info(f"優化完成. 最佳分數: {self.best_score:.4f}")
        logging.info(f"最佳參數: {self.best_params}")
        
        return {
            'best_params': self.best_params,
            'best_score': self.best_score,
            'cv_results': grid_search.cv_results_
        }
```

```python
def train_model(stock_data: pd.DataFrame, 
               model_type: str = "random_forest",
               optimize: bool = True) -> Dict[str, Any]:
    """
    模型訓練主流程
    
    Args:
        stock_data: 含有03_feature_generator生成特徵的DataFrame
        model_type: 模型類型
        optimize: 是否進行參數優化
        
    Returns:
        訓練結果字典
    """
    try:
        logging.info(f"開始訓練 {model_type} 模型...")
        
        # 使用03生成的特徵
        feature_columns = [
            '趨勢強度', '均線糾結度', '波動率', '振幅', '量能趨勢',
            '量比', 'RSI_動能', 'MACD_動能', 'RSI', 'MACD',
            'KD_差值', '本益比_相對值', '技術綜合評分'
        ]
        
        # 準備數據
        X = stock_data[feature_columns].values
        # 預測目標: 1表示隔日上漲,0表示下跌
        y = (stock_data['收盤價'].shift(-1) > stock_data['收盤價']).astype(int).values[:-1]
        X = X[:-1]  # 移除最後一行
        
        # 分割數據
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, shuffle=False
        )
        
        # 參數優化
        if optimize:
            optimizer = ModelOptimizer(
                model_type,
                getattr(ModelConfig, f"{model_type.upper()}_PARAM_GRID")
            )
            opt_results = optimizer.optimize(X_train, y_train)
            params = opt_results['best_params']
        else:
            params = getattr(ModelConfig, f"{model_type.upper()}_PARAMS")
            
        # 訓練模型
        model = ModelFactory.create_model(model_type, params)
        model.train(X_train, y_train)
        
        # 評估模型
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)
        
        # 計算評估指標
        evaluation = ModelEvaluator.evaluate_classification(y_test, y_pred, y_prob)
        
        # 輸出評估報告
        ModelEvaluator.print_evaluation_report(evaluation, model_type)
        ModelEvaluator.plot_confusion_matrix(y_test, y_pred)
        
        return {
            'model': model,
            'evaluation': evaluation,
            'optimization_results': opt_results if optimize else None,
            'test_predictions': {
                'y_true': y_test,
                'y_pred': y_pred,
                'y_prob': y_prob
            },
            'feature_importance': dict(zip(feature_columns, 
                model.model.feature_importances_)) if hasattr(model.model, 'feature_importances_') else None
        }
        
    except Exception as e:
        logging.error(f"訓練過程中出錯: {str(e)}")
        raise
```
