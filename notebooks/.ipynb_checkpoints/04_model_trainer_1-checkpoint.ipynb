```python
# 系統相關
import sys
import os
import argparse
from dataclasses import dataclass
import shutil
import time
from abc import ABC, abstractmethod

# 基礎科學計算
import numpy as np
import pandas as pd
from scipy import stats

# 機器學習
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix, f1_score,
                           precision_score, recall_score, roc_auc_score)
from sklearn.model_selection import (GridSearchCV, KFold, cross_val_score,
                                   train_test_split)
import xgboost as xgb

# 視覺化
import matplotlib.pyplot as plt
import seaborn as sns

# 輔助工具
from tqdm import tqdm  # 進度條顯示
```

```python
class LoggerConfigurator:
    """日誌配置管理器"""
    
    DEFAULT_FORMAT = '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
    DEFAULT_LEVEL = logging.INFO
    
    @classmethod
    def setup(cls, 
              log_path: Optional[Path] = None,
              level: int = DEFAULT_LEVEL,
              format_str: str = DEFAULT_FORMAT) -> None:
        """設置日誌系統
        
        Args:
            log_path: 日誌文件路徑,None則只輸出到控制台
            level: 日誌等級
            format_str: 日誌格式
        """
        handlers = [logging.StreamHandler()]
        
        if log_path:
            log_path.parent.mkdir(parents=True, exist_ok=True)
            handlers.append(logging.FileHandler(
                filename=log_path,
                encoding='utf-8'
            ))
        
        logging.basicConfig(
            level=level,
            format=format_str,
            handlers=handlers
        )
        
    @classmethod
    def get_logger(cls, name: str) -> logging.Logger:
        """獲取命名的日誌器"""
        return logging.getLogger(name)
```

```python
class ModelPaths:
    """模型相關路徑管理"""
    BASE_DIR = Path("D:/Min/Python/Project/FA_Data")
    MODEL_DIR = BASE_DIR / "models"  # 儲存訓練好的模型
    META_DIR = BASE_DIR / "meta_data"  # 元數據
    FEATURE_DIR = BASE_DIR / "features"  # 特徵數據
    RESULTS_DIR = BASE_DIR / "results"  # 預測結果

    def __init__(self):
        """初始化時確保所有必要目錄存在"""
        self._ensure_directories()
    
    def _ensure_directories(self):
        """確保所有必要的目錄存在"""
        directories = [
            self.MODEL_DIR,
            self.META_DIR,
            self.FEATURE_DIR,
            self.RESULTS_DIR
        ]
        
        for directory in directories:
            directory.mkdir(parents=True, exist_ok=True)
            
    @classmethod
    def initialize(cls):
        """創建一個實例來確保目錄存在"""
        return cls()
            
    @staticmethod
    def get_model_path(stock_id: str, model_type: str) -> Path:
        """獲取模型儲存路徑"""
        return ModelPaths.MODEL_DIR / f"{stock_id}_{model_type}_model.pkl"
    
    @staticmethod
    def get_result_path(stock_id: str, model_type: str) -> Path:
        """獲取結果儲存路徑"""
        return ModelPaths.RESULTS_DIR / f"{stock_id}_{model_type}_results.json"
```

```python
class BaseModel(ABC):
    """模型基礎類"""
    
    def __init__(self):
        self.model = None
        self.train_time = None
        self.logger = LoggerConfigurator.get_logger(self.__class__.__name__)
        
    @abstractmethod
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        """訓練模型"""
        pass
        
    @abstractmethod
    def predict(self, X: np.ndarray) -> np.ndarray:
        """預測標籤"""
        pass
        
    @abstractmethod
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """預測概率"""
        pass
        
    def save(self, path: Path) -> None:
        """保存模型"""
        try:
            if self.model is None:
                raise ValueError("沒有可保存的模型")
                
            self.logger.info(f"開始保存模型到: {path}")
            path.parent.mkdir(parents=True, exist_ok=True)
            
            with open(path, 'wb') as f:
                pickle.dump({
                    'model': self.model,
                    'train_time': self.train_time,
                    'model_type': self.__class__.__name__
                }, f)
                
            self.logger.info("模型保存完成")
            
        except Exception as e:
            self.logger.error(f"保存模型時出錯: {str(e)}")
            raise
            
    @classmethod
    def load(cls, path: Path) -> 'BaseModel':
        """載入模型"""
        logger = LoggerConfigurator.get_logger(cls.__name__)
        try:
            if not path.exists():
                raise FileNotFoundError(f"找不到模型文件: {path}")
                
            logger.info(f"開始從 {path} 載入模型")
            
            with open(path, 'rb') as f:
                data = pickle.load(f)
                
            # 根據保存的模型類型創建對應的實例
            model_type = data.get('model_type', cls.__name__)
            instance = ModelFactory.create_model(model_type)
            instance.model = data['model']
            instance.train_time = data.get('train_time')
            
            logger.info("模型載入完成")
            return instance
            
        except Exception as e:
            logger.error(f"載入模型時出錯: {str(e)}")
            raise
            
    def get_feature_importance(self, feature_names: List[str]) -> Dict[str, float]:
        """獲取特徵重要性"""
        try:
            if self.model is None:
                raise ValueError("模型尚未訓練")
                
            if not hasattr(self.model, 'feature_importances_'):
                return {}
                
            self.logger.info("開始計算特徵重要性")
            importance_dict = dict(zip(feature_names, self.model.feature_importances_))
            
            # 標準化特徵重要性
            total = sum(importance_dict.values())
            if total > 0:
                importance_dict = {k: v/total for k, v in importance_dict.items()}
                
            return importance_dict
            
        except Exception as e:
            self.logger.error(f"獲取特徵重要性時出錯: {str(e)}")
            return {}
```

```python
class ModelFactory:
    """模型工廠類"""
    
    _MODEL_REGISTRY = {}
    
    @classmethod
    def register_model(cls, name: str, model_class: type):
        """註冊模型類"""
        cls._MODEL_REGISTRY[name.lower()] = model_class
        
    @classmethod
    def create_model(cls, model_type: str, params: Dict[str, Any] = None) -> BaseModel:
        """創建模型實例"""
        if isinstance(model_type, str):
            model_type = model_type.lower()
            
        if model_type not in cls._MODEL_REGISTRY:
            raise ValueError(f"未註冊的模型類型: {model_type}")
            
        model_class = cls._MODEL_REGISTRY[model_type]
        return model_class(params)
        
    @classmethod
    def get_registered_models(cls) -> List[str]:
        """獲取所有已註冊的模型類型"""
        return list(cls._MODEL_REGISTRY.keys())
```

```python
@dataclass
class EvaluationMetrics:
    """評估指標數據類"""
    accuracy: float
    precision: float
    recall: float
    f1: float
    auc: Optional[float] = None
    confusion_matrix: Optional[np.ndarray] = None

class RandomForestModel(BaseModel):
    """隨機森林模型"""
    
    def __init__(self, params: Optional[Dict] = None):
        super().__init__()
        default_params = {
            'n_estimators': 100,
            'max_depth': 5,
            'random_state': 42,
            'n_jobs': -1
        }
        self.params = default_params if params is None else {**default_params, **params}
        self.model = RandomForestClassifier(**self.params)
        
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        try:
            self.logger.info(f"開始訓練隨機森林模型，數據形狀: X={X.shape}, y={y.shape}")
            start_time = datetime.now()
            self.model.fit(X, y)
            self.train_time = (datetime.now() - start_time).total_seconds()
            self.logger.info(f"模型訓練完成，耗時: {self.train_time:.2f}秒")
        except Exception as e:
            self.logger.error(f"訓練過程出錯: {str(e)}")
            raise
            
    def predict(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict(X)
        
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict_proba(X)

class XGBoostModel(BaseModel):
    """XGBoost模型"""
    
    def __init__(self, params: Optional[Dict] = None):
        super().__init__()
        default_params = {
            'max_depth': 6,
            'learning_rate': 0.1,
            'objective': 'binary:logistic',
            'random_state': 42
        }
        self.params = default_params if params is None else {**default_params, **params}
        self.model = xgb.XGBClassifier(**self.params)
        
    def train(self, X: np.ndarray, y: np.ndarray) -> None:
        try:
            self.logger.info(f"開始訓練XGBoost模型，數據形狀: X={X.shape}, y={y.shape}")
            start_time = datetime.now()
            self.model.fit(X, y)
            self.train_time = (datetime.now() - start_time).total_seconds()
            self.logger.info(f"模型訓練完成，耗時: {self.train_time:.2f}秒")
        except Exception as e:
            self.logger.error(f"訓練過程出錯: {str(e)}")
            raise
            
    def predict(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict(X)
        
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        return self.model.predict_proba(X)

class ModelEvaluation:
    """模型評估類"""
    
    def __init__(self):
        self.logger = LoggerConfigurator.get_logger(self.__class__.__name__)
        
    def evaluate_model(self, 
                      model: BaseModel,
                      X_test: np.ndarray,
                      y_test: np.ndarray) -> EvaluationMetrics:
        """評估模型性能"""
        try:
            y_pred = model.predict(X_test)
            y_prob = model.predict_proba(X_test)
            
            metrics = EvaluationMetrics(
                accuracy=accuracy_score(y_test, y_pred),
                precision=precision_score(y_test, y_pred),
                recall=recall_score(y_test, y_pred),
                f1=f1_score(y_test, y_pred),
                auc=roc_auc_score(y_test, y_prob[:, 1]),
                confusion_matrix=confusion_matrix(y_test, y_pred)
            )
            
            return metrics
            
        except Exception as e:
            self.logger.error(f"評估模型時發生錯誤: {str(e)}")
            raise
            
    def check_model_stability(self,
                            model: BaseModel,
                            X: np.ndarray,
                            y: np.ndarray,
                            n_splits: int = 5) -> Dict[str, float]:
        """檢查模型穩定性"""
        try:
            kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
            fold_metrics = []
            
            for train_idx, val_idx in kf.split(X):
                X_train, X_val = X[train_idx], X[val_idx]
                y_train, y_val = y[train_idx], y[val_idx]
                
                model_fold = copy.deepcopy(model)
                model_fold.train(X_train, y_train)
                metrics = self.evaluate_model(model_fold, X_val, y_val)
                fold_metrics.append(metrics)
            
            # 計算穩定性指標
            stability_metrics = {
                'mean_accuracy': np.mean([m.accuracy for m in fold_metrics]),
                'std_accuracy': np.std([m.accuracy for m in fold_metrics]),
                'mean_f1': np.mean([m.f1 for m in fold_metrics]),
                'std_f1': np.std([m.f1 for m in fold_metrics]),
                'cv_score': np.std([m.accuracy for m in fold_metrics]) / 
                           np.mean([m.accuracy for m in fold_metrics])
            }
            
            return stability_metrics
            
        except Exception as e:
            self.logger.error(f"檢查模型穩定性時發生錯誤: {str(e)}")
            raise
            
    def evaluate_feature_stability(self,
                                model: BaseModel,
                                X: np.ndarray,
                                y: np.ndarray,
                                feature_names: List[str],
                                n_splits: int = 5) -> Dict[str, Dict[str, float]]:
        """評估特徵重要性的穩定性"""
        try:
            kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
            feature_importance_scores = {feature: [] for feature in feature_names}
            
            for train_idx, _ in kf.split(X):
                X_fold, y_fold = X[train_idx], y[train_idx]
                model_fold = copy.deepcopy(model)
                model_fold.train(X_fold, y_fold)
                
                importances = model_fold.get_feature_importance(feature_names)
                for feature, score in importances.items():
                    feature_importance_scores[feature].append(score)
            
            # 計算每個特徵的穩定性指標
            stability_results = {}
            for feature, scores in feature_importance_scores.items():
                if scores:  # 確保有分數
                    stability_results[feature] = {
                        'mean': float(np.mean(scores)),
                        'std': float(np.std(scores)),
                        'cv': float(np.std(scores) / np.mean(scores)) if np.mean(scores) != 0 else float('inf')
                    }
            
            return stability_results
            
        except Exception as e:
            self.logger.error(f"評估特徵穩定性時發生錯誤: {str(e)}")
            raise

# 註冊模型
ModelFactory.register_model('random_forest', RandomForestModel)
ModelFactory.register_model('xgboost', XGBoostModel)
```

```python
@dataclass
class FeatureConfig:
    """特徵工程配置"""
    required_features: List[str] = None
    
    def __post_init__(self):
        if self.required_features is None:
            self.required_features = [
                '趨勢強度', '振幅', '量能趨勢', '量比', 'RSI', 'MACD',  # 基礎特徵
                '均線糾結度', '波動率', 'RSI_動能', 'MACD_動能',        # 技術特徵
                'KD_差值', '本益比_相對值', '技術綜合評分'             # 綜合指標
            ]

class FeatureEngineer:
    """特徵工程類"""
    
    def __init__(self, config: Optional[FeatureConfig] = None):
        self.config = config or FeatureConfig()
        self.logger = LoggerConfigurator.get_logger(self.__class__.__name__)
        
    def process(self, df: pd.DataFrame) -> Tuple[np.ndarray, np.ndarray, List[str]]:
        """處理特徵工程"""
        try:
            self.logger.info("開始特徵工程處理")
            
            # 驗證數據
            self._validate_data(df)
            
            # 清理數據
            df = self._clean_data(df)
            
            # 生成特徵
            X, y, feature_names = self._generate_features(df)
            
            self.logger.info(f"特徵工程完成，特徵形狀: {X.shape}, 標籤形狀: {y.shape}")
            return X, y, feature_names
            
        except Exception as e:
            self.logger.error(f"特徵工程處理失敗: {str(e)}")
            raise
            
    def _validate_data(self, df: pd.DataFrame) -> None:
        """驗證數據完整性"""
        # 檢查數據量
        if len(df) < 30:
            raise ValueError("數據量不足，需要至少30天的數據")
            
        # 檢查必要特徵
        missing_features = [f for f in self.config.required_features 
                          if f not in df.columns]
        if missing_features:
            raise ValueError(f"缺少以下特徵: {missing_features}")
            
        # 檢查並記錄缺失值
        null_stats = df[self.config.required_features].isnull().sum()
        if null_stats.any():
            self.logger.warning("數據存在缺失值:")
            for feature, null_count in null_stats[null_stats > 0].items():
                self.logger.warning(f"{feature}: {null_count} 個缺失值")
                
    def _clean_data(self, df: pd.DataFrame) -> pd.DataFrame:
        """清理數據"""
        df = df.copy()
        
        # 清理重複列名
        df = self._clean_column_names(df)
        
        # 處理缺失值
        df = self._handle_missing_values(df)
        
        # 數據類型轉換
        df = self._convert_data_types(df)
        
        return df
    
    def _clean_column_names(self, df: pd.DataFrame) -> pd.DataFrame:
        """清理列名"""
        df = df.copy()
        columns_to_clean = [col for col in df.columns 
                          if col.endswith('_x') or col.endswith('_y')]
        
        for col in columns_to_clean:
            base_col = col[:-2]
            if f"{base_col}_x" in df.columns:
                df[base_col] = df[f"{base_col}_x"]
                df.drop([f"{base_col}_x", f"{base_col}_y"], 
                       axis=1, inplace=True, errors='ignore')
        return df
    
    def _handle_missing_values(self, df: pd.DataFrame) -> pd.DataFrame:
        """處理缺失值"""
        df = df.copy()
        
        for feature in self.config.required_features:
            if feature not in df.columns:
                self.logger.warning(f"缺少特徵 {feature}，將被設為0")
                df[feature] = 0
            else:
                # 使用前向填充和後向填充
                df[feature] = df[feature].fillna(method='ffill').fillna(method='bfill')
                
                # 如果仍有缺失值，使用均值填充
                if df[feature].isnull().any():
                    feature_mean = df[feature].mean()
                    df[feature] = df[feature].fillna(feature_mean)
                    self.logger.info(f"特徵 {feature} 的剩餘缺失值使用均值 {feature_mean:.4f} 填充")
        
        return df
    
    def _convert_data_types(self, df: pd.DataFrame) -> pd.DataFrame:
        """轉換數據類型"""
        df = df.copy()
        
        # 確保數值型特徵
        for feature in self.config.required_features:
            if feature in df.columns:
                df[feature] = pd.to_numeric(df[feature], errors='coerce')
        
        return df
        
    def _generate_features(self, df: pd.DataFrame) -> Tuple[np.ndarray, np.ndarray, List[str]]:
        """生成特徵"""
        # 準備特徵矩陣
        X = df[self.config.required_features].values
        
        # 生成標籤（預測下一天的漲跌）
        y = (df['收盤價'].shift(-1) > df['收盤價']).astype(int).values[:-1]
        X = X[:-1]  # 移除最後一行以匹配標籤長度
        
        return X, y, self.config.required_features
```

```python
class ResultManager:
    """結果管理類"""
    
    def __init__(self, base_path: Path):
        self.base_path = base_path
        self.logger = LoggerConfigurator.get_logger(self.__class__.__name__)
        self.results_dir = base_path / "results"
        self.backup_dir = base_path / "backup" / "results"
        
        # 確保目錄存在
        self.results_dir.mkdir(parents=True, exist_ok=True)
        self.backup_dir.mkdir(parents=True, exist_ok=True)
        
    def save_results(self, 
                    stock_id: str,
                    model_type: str,
                    results: Dict[str, Any],
                    backup: bool = True) -> None:
        """保存評估結果"""
        try:
            # 添加元數據
            results['metadata'] = {
                'stock_id': stock_id,
                'model_type': model_type,
                'timestamp': datetime.now().isoformat(),
                'version': '1.0'
            }
            
            # 保存結果
            result_path = self.results_dir / f"{stock_id}_{model_type}_results.json"
            self._save_json(result_path, results)
            
            # 創建備份
            if backup:
                self._create_backup(result_path)
                
            self.logger.info(f"結果已保存到: {result_path}")
            
        except Exception as e:
            self.logger.error(f"保存結果時發生錯誤: {str(e)}")
            raise
            
    def load_results(self, stock_id: str, model_type: str) -> Optional[Dict]:
        """載入評估結果"""
        try:
            result_path = self.results_dir / f"{stock_id}_{model_type}_results.json"
            
            if not result_path.exists():
                self.logger.warning(f"找不到結果文件: {result_path}")
                return None
                
            return self._load_json(result_path)
            
        except Exception as e:
            self.logger.error(f"載入結果時發生錯誤: {str(e)}")
            return None
            
    def _save_json(self, path: Path, data: Dict) -> None:
        """保存JSON文件"""
        with open(path, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
            
    def _load_json(self, path: Path) -> Dict:
        """載入JSON文件"""
        with open(path, 'r', encoding='utf-8') as f:
            return json.load(f)
            
    def _create_backup(self, file_path: Path) -> None:
        """創建備份"""
        backup_time = datetime.now().strftime('%Y%m%d_%H%M%S')
        backup_path = self.backup_dir / f"{file_path.stem}_{backup_time}{file_path.suffix}"
        shutil.copy2(file_path, backup_path)
        
        # 清理舊備份
        self._cleanup_old_backups(file_path.stem)
        
    def _cleanup_old_backups(self, result_prefix: str, keep_days: int = 7) -> None:
        """清理舊備份"""
        current_time = datetime.now().timestamp()
        
        for backup_file in self.backup_dir.glob(f"{result_prefix}_*"):
            if (current_time - backup_file.stat().st_mtime) > (keep_days * 86400):
                backup_file.unlink()
                self.logger.info(f"已刪除舊備份: {backup_file}")
                
    def get_latest_results(self, stock_id: str = None) -> Dict[str, Dict]:
        """獲取最新結果"""
        try:
            results = {}
            pattern = f"{stock_id}_*_results.json" if stock_id else "*_results.json"
            
            for result_file in self.results_dir.glob(pattern):
                data = self._load_json(result_file)
                if 'metadata' in data:
                    key = f"{data['metadata']['stock_id']}_{data['metadata']['model_type']}"
                    results[key] = data
                    
            return results
            
        except Exception as e:
            self.logger.error(f"獲取最新結果時發生錯誤: {str(e)}")
            return {}
```

```python
class ModelOptimizer:
    """模型優化器"""
    
    def __init__(self, model_type: str, param_grid: Dict[str, List[Any]]):
        self.model_type = model_type
        self.param_grid = param_grid
        self.best_params = None
        self.best_score = None
        
    def optimize(self, X: np.ndarray, y: np.ndarray, 
                cv: int = 5, scoring: str = 'accuracy') -> Dict[str, Any]:
        """
        執行參數優化
        
        Args:
            X: 特徵矩陣
            y: 標籤向量
            cv: 交叉驗證折數
            scoring: 評分標準
            
        Returns:
            優化結果字典
        """
        logging.info(f"開始對 {self.model_type} 進行參數優化...")
        
        grid_search = GridSearchCV(
            ModelFactory.create_model(self.model_type).model,
            self.param_grid,
            cv=cv,
            scoring=scoring,
            n_jobs=-1,
            verbose=1
        )
        
        grid_search.fit(X, y)
        
        self.best_params = grid_search.best_params_
        self.best_score = grid_search.best_score_
        
        logging.info(f"優化完成. 最佳分數: {self.best_score:.4f}")
        logging.info(f"最佳參數: {self.best_params}")
        
        return {
            'best_params': self.best_params,
            'best_score': self.best_score,
            'cv_results': grid_search.cv_results_
        }
```

```python
@dataclass
class TrainingConfig:
    """訓練配置類"""
    base_path: Path
    test_size: float = 0.2
    optimize: bool = True
    cv_folds: int = 5
    random_state: int = 42
    
    # 模型參數網格
    param_grids: Dict[str, Dict] = None
    
    def __post_init__(self):
        if self.param_grids is None:
            self.param_grids = {
                "random_forest": {
                    'n_estimators': [300, 400, 500],
                    'max_depth': [3, 4, 5, 6],
                    'min_samples_split': [8, 10, 12],
                    'min_samples_leaf': [2, 4],
                    'max_features': ['sqrt', None],
                    'class_weight': [None, 'balanced']
                },
                "xgboost": {
                    'n_estimators': [100, 200, 300],
                    'max_depth': [3, 4, 5, 6],
                    'learning_rate': [0.01, 0.05, 0.1],
                    'subsample': [0.8, 0.9, 1.0],
                    'colsample_bytree': [0.8, 0.9, 1.0],
                    'min_child_weight': [1, 3, 5]
                }
            }

class ModelTrainer:
    """模型訓練管理類"""
    
    def __init__(self, stock_id: str, config: Optional[TrainingConfig] = None):
        """初始化
        
        Args:
            stock_id: 股票代碼
            config: 訓練配置
        """
        self.stock_id = stock_id
        self.config = config or TrainingConfig(base_path=Path("D:/Min/Python/Project/FA_Data"))
        self.logger = LoggerConfigurator.get_logger(self.__class__.__name__)
        
        # 初始化組件
        self.feature_engineer = FeatureEngineer()
        self.evaluator = ModelEvaluation()
        self.result_manager = ResultManager(self.config.base_path)
        
    def train_and_evaluate(self, 
                          model_types: List[str] = ["random_forest", "xgboost"]) -> Dict[str, Any]:
        """執行完整的訓練和評估流程"""
        try:
            self.logger.info(f"開始訓練股票 {self.stock_id} 的模型")
            
            # 1. 讀取和準備數據
            stock_data = self._load_data()
            X, y, feature_names = self.feature_engineer.process(stock_data)
            
            # 2. 分割數據
            X_train, X_test, y_train, y_test = self._split_data(X, y)
            
            # 3. 訓練和評估每個模型
            models = []
            individual_results = {}
            
            for model_type in model_types:
                model_result = self._train_single_model(
                    model_type,
                    X_train, y_train,
                    X_test, y_test,
                    feature_names
                )
                models.append(model_result['model'])
                individual_results[model_type] = model_result['evaluation']
            
            # 4. 創建和評估集成模型
            ensemble_result = self._create_ensemble_model(
                models,
                X_train, y_train,
                X_test, y_test,
                feature_names
            )
            
            # 5. 整合所有結果
            final_results = {
                'ensemble': ensemble_result,
                'individual_models': individual_results,
                'feature_names': feature_names,
                'training_info': {
                    'data_shape': X.shape,
                    'test_size': self.config.test_size,
                    'training_time': datetime.now().isoformat()
                }
            }
            
            # 6. 保存結果
            self.result_manager.save_results(
                self.stock_id,
                'ensemble',
                final_results
            )
            
            self.logger.info("模型訓練和評估完成")
            return final_results
            
        except Exception as e:
            self.logger.error(f"訓練和評估過程出錯: {str(e)}")
            raise
            
    def _load_data(self) -> pd.DataFrame:
        """載入股票數據"""
        try:
            data_path = self.config.base_path / "meta_data" / "enhanced_features.csv"
            if not data_path.exists():
                raise FileNotFoundError(f"找不到數據文件: {data_path}")
                
            df = pd.read_csv(data_path, dtype={'證券代號': str})
            stock_data = df[df['證券代號'] == self.stock_id].copy()
            
            if len(stock_data) == 0:
                raise ValueError(f"找不到股票 {self.stock_id} 的數據")
                
            return stock_data
            
        except Exception as e:
            self.logger.error(f"載入數據時出錯: {str(e)}")
            raise
            
    def _split_data(self, X: np.ndarray, y: np.ndarray) -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
        """分割訓練和測試數據"""
        return train_test_split(
            X, y,
            test_size=self.config.test_size,
            random_state=self.config.random_state,
            shuffle=False  # 時間序列數據不打亂
        )
        
    def _train_single_model(self,
                          model_type: str,
                          X_train: np.ndarray,
                          y_train: np.ndarray,
                          X_test: np.ndarray,
                          y_test: np.ndarray,
                          feature_names: List[str]) -> Dict[str, Any]:
        """訓練單個模型"""
        try:
            self.logger.info(f"開始訓練 {model_type} 模型")
            
            # 創建模型
            model = ModelFactory.create_model(model_type)
            
            # 參數優化
            if self.config.optimize:
                param_grid = self.config.param_grids.get(model_type, {})
                if param_grid:
                    self._optimize_parameters(model, param_grid, X_train, y_train)
            
            # 訓練模型
            model.train(X_train, y_train)
            
            # 評估模型
            evaluation_metrics = self.evaluator.evaluate_model(model, X_test, y_test)
            stability_metrics = self.evaluator.check_model_stability(
                model, X_train, y_train, self.config.cv_folds
            )
            feature_stability = self.evaluator.evaluate_feature_stability(
                model, X_train, y_train, feature_names, self.config.cv_folds
            )
            
            return {
                'model': model,
                'evaluation': {
                    'metrics': vars(evaluation_metrics),
                    'stability': stability_metrics,
                    'feature_stability': feature_stability,
                    'feature_importance': model.get_feature_importance(feature_names)
                }
            }
            
        except Exception as e:
            self.logger.error(f"訓練 {model_type} 模型時出錯: {str(e)}")
            raise
            
    def _optimize_parameters(self,
                           model: BaseModel,
                           param_grid: Dict[str, List[Any]],
                           X: np.ndarray,
                           y: np.ndarray) -> None:
        """優化模型參數"""
        try:
            from sklearn.model_selection import GridSearchCV
            
            grid_search = GridSearchCV(
                model.model,
                param_grid,
                cv=self.config.cv_folds,
                n_jobs=-1,
                scoring='accuracy'
            )
            
            grid_search.fit(X, y)
            model.model.set_params(**grid_search.best_params_)
            self.logger.info(f"最佳參數: {grid_search.best_params_}")
            
        except Exception as e:
            self.logger.error(f"參數優化時出錯: {str(e)}")
            raise
            
    def _create_ensemble_model(self,
                             models: List[BaseModel],
                             X_train: np.ndarray,
                             y_train: np.ndarray,
                             X_test: np.ndarray,
                             y_test: np.ndarray,
                             feature_names: List[str]) -> Dict[str, Any]:
        """創建和評估集成模型"""
        try:
            ensemble = EnsembleModel(models)
            ensemble.train(X_train, y_train)
            
            evaluation_metrics = self.evaluator.evaluate_model(ensemble, X_test, y_test)
            stability_metrics = self.evaluator.check_model_stability(
                ensemble, X_train, y_train, self.config.cv_folds
            )
            feature_stability = self.evaluator.evaluate_feature_stability(
                ensemble, X_train, y_train, feature_names, self.config.cv_folds
            )
            
            return {
                'metrics': vars(evaluation_metrics),
                'stability': stability_metrics,
                'feature_stability': feature_stability,
                'feature_importance': ensemble.get_feature_importance(feature_names)
            }
            
        except Exception as e:
            self.logger.error(f"創建集成模型時出錯: {str(e)}")
            raise
```

```python
def setup_logging(log_path: Optional[str] = None) -> None:
    """設置日誌系統"""
    if log_path:
        log_file = Path(log_path)
        log_file.parent.mkdir(parents=True, exist_ok=True)
        LoggerConfigurator.setup(log_file)
    else:
        LoggerConfigurator.setup()

def parse_arguments() -> argparse.Namespace:
    """解析命令行參數"""
    parser = argparse.ArgumentParser(description='台股模型訓練系統')
    parser.add_argument('--config', type=str, help='配置文件路徑')
    parser.add_argument('--stocks', type=str, nargs='+', help='要處理的股票代碼列表')
    parser.add_argument('--log', type=str, help='日誌文件路徑')
    return parser.parse_args()

def main():
    """主程式入口"""
    try:
        # 解析參數
        args = parse_arguments()
        
        # 設置日誌
        setup_logging(args.log)
        logger = LoggerConfigurator.get_logger("main")
        logger.info("開始執行模型訓練")
        
        # 創建協調器
        config_path = Path(args.config) if args.config else None
        orchestrator = TrainingOrchestrator(config_path)
        
        # 執行訓練
        results = orchestrator.train_all_stocks(args.stocks)
        
        logger.info("模型訓練完成")
        return 0
        
    except Exception as e:
        logger.error(f"程式執行失敗: {str(e)}")
        return 1
```
