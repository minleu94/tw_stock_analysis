```python
import numpy as np
import pandas as pd
from dataclasses import dataclass
from typing import Dict, List, Optional, Tuple
from datetime import datetime
import logging
from pathlib import Path
import json
```

```python
@dataclass
class PortfolioConfig:
    """投資組合配置參數"""
    # 基礎路徑設定
    BASE_DIR: str = "D:/Min/Python/Project/FA_Data"
    META_DATA_DIR: str = "meta_data"
    PORTFOLIO_DIR: str = "portfolio"
    LOG_DIR: str = "logs"
    
    # 投資組合參數
    PORTFOLIO_PARAMS: Dict = {
        'max_stock_weight': 0.2,  # 單一股票最大權重
        'min_stock_weight': 0.02,  # 單一股票最小權重
        'max_industry_weight': 0.35,  # 單一產業最大權重
        'risk_free_rate': 0.015,  # 無風險利率
        'rebalance_threshold': 0.1,  # 再平衡閾值
        'stop_loss_threshold': -0.1,  # 停損點
        'take_profit_threshold': 0.2,  # 獲利點
    }
    
    # 風險控制參數
    RISK_PARAMS: Dict = {
        'max_portfolio_volatility': 0.25,  # 最大組合波動率
        'max_drawdown_threshold': -0.15,  # 最大回撤限制
        'var_confidence_level': 0.95,  # VaR信心水準
        'position_limit': 20,  # 最大持股數量
    }
```

```python
class PortfolioManager:
    """投資組合管理核心類"""
    
    def __init__(self, config: PortfolioConfig = None):
        self.config = config or PortfolioConfig()
        self.setup_logging()
        self.portfolio: Dict = {}  # 當前持倉
        self.history: List = []    # 歷史交易記錄
        self.performance: Dict = {} # 績效指標
        self._initialize_directories()
    
    def setup_logging(self):
        """設置日誌系統"""
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(f"{self.config.BASE_DIR}/{self.config.LOG_DIR}/portfolio.log", 
                                  encoding='utf-8'),
                logging.StreamHandler()
            ]
        )
        self.logger = logging.getLogger('PortfolioManager')
    
    def _initialize_directories(self):
        """初始化必要目錄"""
        for dir_name in [self.config.META_DATA_DIR, self.config.PORTFOLIO_DIR, self.config.LOG_DIR]:
            directory = Path(f"{self.config.BASE_DIR}/{dir_name}")
            directory.mkdir(parents=True, exist_ok=True)
    
    def calculate_portfolio_metrics(self) -> Dict:
        """計算投資組合指標"""
        try:
            metrics = {
                'total_value': self._calculate_total_value(),
                'total_return': self._calculate_total_return(),
                'volatility': self._calculate_portfolio_volatility(),
                'sharpe_ratio': self._calculate_sharpe_ratio(),
                'max_drawdown': self._calculate_max_drawdown(),
                'value_at_risk': self._calculate_var(),
                'industry_weights': self._calculate_industry_weights(),
                'position_concentration': self._calculate_position_concentration()
            }
            self.performance = metrics
            return metrics
        except Exception as e:
            self.logger.error(f"計算投資組合指標時發生錯誤: {str(e)}")
            return {}
    
    def _calculate_total_value(self) -> float:
        """計算投資組合總價值"""
        return sum(position['shares'] * position['current_price'] 
                  for position in self.portfolio.values())
    
    def _calculate_total_return(self) -> float:
        """計算總報酬率"""
        if not self.history:
            return 0.0
        initial_value = self.history[0]['portfolio_value']
        current_value = self._calculate_total_value()
        return (current_value - initial_value) / initial_value
    
    def _calculate_portfolio_volatility(self) -> float:
        """計算投資組合波動率"""
        if not self.history:
            return 0.0
        returns = pd.Series([record['daily_return'] for record in self.history])
        return returns.std() * np.sqrt(252)  # 年化波動率
    
    def _calculate_sharpe_ratio(self) -> float:
        """計算夏普比率"""
        if not self.history:
            return 0.0
        returns = pd.Series([record['daily_return'] for record in self.history])
        excess_returns = returns - self.config.PORTFOLIO_PARAMS['risk_free_rate'] / 252
        return np.sqrt(252) * excess_returns.mean() / returns.std()
    
    def check_rebalance_signals(self) -> List[Dict]:
        """檢查是否需要再平衡"""
        signals = []
        
        # 檢查個股權重
        for stock_id, position in self.portfolio.items():
            weight = position['market_value'] / self._calculate_total_value()
            
            if weight > self.config.PORTFOLIO_PARAMS['max_stock_weight']:
                signals.append({
                    'stock_id': stock_id,
                    'action': 'reduce',
                    'reason': 'exceed_max_weight',
                    'current_weight': weight,
                    'target_weight': self.config.PORTFOLIO_PARAMS['max_stock_weight']
                })
            
            # 檢查停損點
            returns = position['current_price'] / position['avg_cost'] - 1
            if returns < self.config.PORTFOLIO_PARAMS['stop_loss_threshold']:
                signals.append({
                    'stock_id': stock_id,
                    'action': 'sell',
                    'reason': 'stop_loss',
                    'current_return': returns
                })
            
            # 檢查獲利點
            if returns > self.config.PORTFOLIO_PARAMS['take_profit_threshold']:
                signals.append({
                    'stock_id': stock_id,
                    'action': 'sell',
                    'reason': 'take_profit',
                    'current_return': returns
                })
        
        return signals
    
    def execute_rebalance(self, signals: List[Dict]) -> bool:
        """執行再平衡操作"""
        try:
            for signal in signals:
                if signal['action'] == 'reduce':
                    self._reduce_position(signal['stock_id'], signal['target_weight'])
                elif signal['action'] == 'sell':
                    self._close_position(signal['stock_id'])
                
                self._record_transaction(signal['stock_id'], signal['action'], signal['reason'])
            
            return True
        except Exception as e:
            self.logger.error(f"執行再平衡時發生錯誤: {str(e)}")
            return False
    
    def _reduce_position(self, stock_id: str, target_weight: float):
        """減少持倉"""
        current_position = self.portfolio[stock_id]
        total_value = self._calculate_total_value()
        target_value = total_value * target_weight
        current_value = current_position['shares'] * current_position['current_price']
        
        if current_value > target_value:
            reduction_ratio = (current_value - target_value) / current_value
            new_shares = current_position['shares'] * (1 - reduction_ratio)
            self.portfolio[stock_id]['shares'] = new_shares
    
    def _close_position(self, stock_id: str):
        """平倉"""
        if stock_id in self.portfolio:
            del self.portfolio[stock_id]
    
    def _record_transaction(self, stock_id: str, action: str, reason: str):
        """記錄交易"""
        transaction = {
            'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'stock_id': stock_id,
            'action': action,
            'reason': reason,
            'portfolio_value': self._calculate_total_value(),
            'daily_return': self._calculate_daily_return()
        }
        self.history.append(transaction)
    
    def save_portfolio_state(self):
        """保存投資組合狀態"""
        try:
            state = {
                'portfolio': self.portfolio,
                'performance': self.performance,
                'last_update': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            }
            
            file_path = Path(f"{self.config.BASE_DIR}/{self.config.PORTFOLIO_DIR}/portfolio_state.json")
            with open(file_path, 'w', encoding='utf-8') as f:
                json.dump(state, f, ensure_ascii=False, indent=2)
                
            self.logger.info("投資組合狀態已保存")
            return True
        except Exception as e:
            self.logger.error(f"保存投資組合狀態時發生錯誤: {str(e)}")
            return False
    
    def load_portfolio_state(self) -> bool:
        """載入投資組合狀態"""
        try:
            file_path = Path(f"{self.config.BASE_DIR}/{self.config.PORTFOLIO_DIR}/portfolio_state.json")
            if not file_path.exists():
                self.logger.warning("找不到投資組合狀態文件")
                return False
                
            with open(file_path, 'r', encoding='utf-8') as f:
                state = json.load(f)
                
            self.portfolio = state['portfolio']
            self.performance = state['performance']
            self.logger.info("投資組合狀態已載入")
            return True
        except Exception as e:
            self.logger.error(f"載入投資組合狀態時發生錯誤: {str(e)}")
            return False
```

```python
# 初始化投資組合管理器
config = PortfolioConfig()
portfolio_manager = PortfolioManager(config)

# 載入現有投資組合
portfolio_manager.load_portfolio_state()

# 檢查再平衡信號
signals = portfolio_manager.check_rebalance_signals()

# 執行再平衡(如果需要)
if signals:
    portfolio_manager.execute_rebalance(signals)

# 計算績效指標
metrics = portfolio_manager.calculate_portfolio_metrics()

# 保存當前狀態
portfolio_manager.save_portfolio_state()
```
